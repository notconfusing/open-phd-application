```python
import pandas
%pylab inline
```

```python
def t(s):
    month = '03' if s[0] == 's' else '08'
    year= '20'+s[1:]
    return pandas.Timestamp('%s/%s' %(month, year))
```

```python
grades = {t('f04'):4.0,
                  t('f05'):3.0,
                  t('s06'):2.0,
                  t('f06'):4.0,
                  t('s07'):4.0,
                t('f07'):3.167,
                t('s08'):4.0,
                t('f08'):4.0,
                t('s09'):3.75,
                t('f09'):3.533,
                t('s10'):3.433,
                t('f10'):2.9,
                t('s11'):2.65,}

units = {t('f04'):1.0,
                  t('f05'):4.0,
                  t('s06'):3.0,
                  t('f06'):3.0,
                  t('s07'):5.0,
                t('f07'):12.0,
                t('s08'):12.0,
                t('f08'):16.0,
                t('s09'):16.0,
                t('f09'):13.0,
                t('s10'):14.0,
                t('f10'):14.0,
                t('s11'):10.0,}
```

```python
gradesdf = pandas.DataFrame(data={'gpa':grades,
                                                                      'units':units})
```

```python
gradesdf.plot(kind='line', figsize=(10,4))
```
